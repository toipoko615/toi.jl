```julia
include("toiforMPS.jl")
using .Toimps
using .randmpo
using ITensors
using Quantics
```

```julia
N=5
D=4
d=8
TTS,q=Toimps.randMPS(N,D,d)
```

```julia
TTO=randmpo.randMPO(N,D,d,q)
```

```julia
cutoff=0.1
QTTMPS,TTch,R=Toimps.MPStoQTTMPS(N,cutoff,TTS)
```

```julia
include("toi.jl")
using .Toi
```

```julia
QTTMPO,MPOin,links,R=Toi.MPOtoQTTMPO(TTO,cutoff)
```

```julia
QTTMPO
```

```julia
inds = [Index(2, "a = $i") for i in 1:N*R]
```

```julia
for n in 1:N*R
    if n==1
        replaceinds!(QTTMPO[n],  QTTMPO[n].tensor.inds[1]=> inds[n])
    else
        replaceinds!(QTTMPO[n],  QTTMPO[n].tensor.inds[2]=> inds[n])
    end
end
QTTMPO
```

```julia
for n in 1:N*R
    if n==1
        replaceinds!(QTTMPS[n],  QTTMPS[n].tensor.inds[1]=> inds[n])
    else
        replaceinds!(QTTMPS[n],  QTTMPS[n].tensor.inds[2]=> inds[n])
    end
end
QTTMPS
```

```julia
res1 = apply(TTO, TTS)
```

```julia
inds=[]
for n in 1:5
    if n==1
        push!(inds,res1[n].tensor.inds[2])
    else
        push!(inds,res1[n].tensor.inds[1])
    end
end
inds
```

```julia
res1v = vec(Array(reduce(*, res1), reverse(inds[1:N])))
```

```julia
res2 = apply(QTTMPO, QTTMPS)
```

```julia
inds=[]
for n in 1:N*R
    if n==1
        push!(inds,res2[n].tensor.inds[2])
    else
        push!(inds,res2[n].tensor.inds[1])
    end
end
inds
```

```julia
res2v = vec(Array(reduce(*, res2), reverse(inds[1:N*R])))
```

```julia
# 相対誤差の計算
relative_errors = [abs(r1 - r2) / r1 for (r1, r2) in zip(res1v, res2v)]

# 相対誤差の平均を求める
mean_relative_error = sum(relative_errors) / length(relative_errors)

println(mean_relative_error)
```
